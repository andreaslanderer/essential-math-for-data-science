## Linear Function

```python
def f(x):
    return 2 * x + 1

x_values = [0, 1, 2, 3]

for x_value in x_values:
    y = f(x_value)
    print(f"f({x_value}) = {y}")
```

```python
from sympy import *

x = symbols("x")
f = 2 * x + 1
plot(f)
```

## Exponential Function

```python
from sympy import *

x = symbols("x")
f = x**2 + 1
plot(f)
```

## Functions with multiple input variables

```python
from sympy import *
from sympy.plotting import plot3d

x, y = symbols("x y")
f = 2 * x + 3 * y
plot3d(f)
```

# Summations

## Example 1

$ \sum \limits _{i=1} ^{5} 2i = (2) * 1 + (2) * 2 + (2) * 3 + (2) * 4 + (2) * 5 = 30 $

```python
summation = sum(2*i for i in range(1,6)) # upper bound is exclusive 
print(f"summation = {summation}")
```

## Example 2

$ \sum \limits _{i=1} ^{n} 10x_{i} $

```python
x = [1, 4, 6, 2]
n = len(x)

summation_1 = sum(10*x[i] for i in range(0, n))
print(f"summation_1 = {summation_1}")
```

## Using the SymPy summation operator

```python
from sympy import *

x = [1, 4, 6, 2]
i, n = symbols("i n")

# Iterate each element i from 1 to n
# Then multiply and sum
# summation = (10)*1 + (10)*2 + (10)*3 + (10)*4 + (10)*5 = 150
summation = Sum(10 * i, (i, 1, n))

# Specify n as 5
# Iterating the numbers 1 through 5
up_to_5 = summation.subs(n, 5)
print(up_to_5.doit())
```
